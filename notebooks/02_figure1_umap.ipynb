```python
import scanpy as sc

from src.config import FIGURES_DIR, PROCESSED_DATA_DIR
from src.plotting import setup_scanpy

setup_scanpy()

adata = sc.read_h5ad(PROCESSED_DATA_DIR / "kidney_loaded.h5ad")
adata
```

```python
# 若已有 UMAP 坐标（官方 h5ad 通常包含），直接绘图
color_key = "cell_type" if "cell_type" in adata.obs.columns else adata.obs.columns[0]

if "X_umap" not in adata.obsm:
  sc.pp.normalize_total(adata, target_sum=1e4)
  sc.pp.log1p(adata)
  sc.pp.highly_variable_genes(adata, n_top_genes=2000)
  sc.pp.pca(adata)
  sc.pp.neighbors(adata)
  sc.tl.umap(adata)

sc.pl.umap(adata, color=color_key, show=False)
FIGURES_DIR.mkdir(parents=True, exist_ok=True)
sc.pl.umap(adata, color=color_key, save="_figure1_umap.png", show=True)
```
