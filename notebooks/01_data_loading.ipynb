```python
import scanpy as sc
from pathlib import Path

from src.config import PROCESSED_DATA_DIR, RAW_DATA_DIR
from src.plotting import setup_scanpy

setup_scanpy()
RAW_DATA_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DATA_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# 将下载的 h5ad 路径填在这里
H5AD_PATH = Path("data/raw/kidney.h5ad")  # 或 Google Drive 路径

if not H5AD_PATH.exists():
    raise FileNotFoundError(
        f"未找到数据文件: {H5AD_PATH}\n"
        "请先从 KPMP / CELLxGENE 下载 h5ad，或运行 00_setup_colab.ipynb 挂载 Drive。"
    )

adata = sc.read_h5ad(H5AD_PATH)
print(adata)
adata
```

```python
# 保存标准化后的对象，供后续 figure notebook 复用
out_path = PROCESSED_DATA_DIR / "kidney_loaded.h5ad"
adata.write_h5ad(out_path)
print("Saved:", out_path)
```
